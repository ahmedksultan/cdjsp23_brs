```python
import json
import math

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sqlalchemy
```

```python
all_data = pd.read_csv('all_cornell.csv')
all_data
```

```python
gym_data = pd.read_csv('all_cornell_gym.csv')
study_data = pd.read_csv('all_cornell_study.csv')
party_data = pd.read_csv('all_cornell_party.csv')
commute_data = pd.read_csv('all_cornell_commute.csv')
sleep_data = pd.read_csv('all_cornell_sleep.csv')


for data in [gym_data, study_data, party_data, commute_data, sleep_data]:
    print('number of songs submitted:', len(data), 'number of users submitted:', data['username'].nunique())
```

```python
def filter_top(pair, count):
    key, value = pair
    return value > count


top_songs = dict((party_data['title'] + ' - ' + party_data['artist']).value_counts())
top_songs = dict((filter(lambda x: filter_top(x, 4), top_songs.items())))

top_songs
```

```python
top_artists = dict(party_data['artist'].value_counts())

top_artists = dict((filter(lambda x: filter_top(x, 10), top_artists.items())))
top_artists
```

```python
names = list(top_songs.keys())
values = list(top_songs.values())

plt.barh(range(len(top_songs)), values, tick_label=names)
plt.gca().invert_yaxis()
plt.show()
```

```python
names = list(top_artists.keys())
values = list(top_artists.values())

plt.barh(range(len(top_artists)), values, tick_label=names)
plt.gca().invert_yaxis()
plt.show()
```
